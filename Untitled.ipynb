```python
class Gen_Sneha(nn.Module):
    def __init__(self, in_size=3, h=5,outsize=14943, p=0.0):
        
        super().__init__()
        self.in_size = in_size
        self.outsize = outsize
        self.h = h
        self.L= h
        
        self.hidden=nn.ModuleList()
        self.Layers = [in_size, *[2**(2*i) for i in range(3,h+2)], outsize]
        print(self.Layers)
        
        for i,(input_size,output_size) in enumerate(zip(self.Layers, self.Layers[1:])):
            if i<self.L-1:
                self.hidden.append(nn.Linear(input_size, output_size, bias=False))
                self.hidden.append(nn.BatchNorm1d(output_size))
                self.hidden.append(nn.ReLU())
                self.hidden.append(nn.Dropout(p=p))
            else:
                self.hidden.append(nn.Linear(input_size, output_size, bias=False))
                self.hidden.append(nn.Tanh())
                
    def forward(self, activation):
        for linear_transform in self.hidden:
            activation= linear_transform(activation)
        return activation
```

```python
class Dis_Sneha(nn.Module):
    def __init__(self, in_size=14943, h=5,outsize=1, p=0.33, Wgan=False):
        
        super().__init__()
        self.in_size = in_size
        self.outsize = outsize
        self.h = h
        self.L= h
       
        self.Wgan=Wgan
        self.hidden=nn.ModuleList()
        self.Layers = [in_size, *[2**(2*i) for i in range(h+1,2,-1)], outsize]
        print(self.Layers)
    
        for i,(input_size,output_size) in enumerate(zip(self.Layers, self.Layers[1:])):
            if i<self.L-1:
                self.hidden.append(nn.Linear(input_size, output_size, bias=False))
                self.hidden.append(nn.BatchNorm1d(output_size))
                self.hidden.append(nn.LeakyReLU(0.2))
                self.hidden.append(nn.Dropout(p=p))
            else:
                self.hidden.append(nn.Linear(input_size, output_size, bias=False))
                if not self.Wgan:
                    self.hidden.append(nn.Sigmoid())
                
    def forward(self, activation):
        for linear_transform in self.hidden:
            activation= linear_transform(activation)
        return activation
```

```python
class Residual_Base(nn.Module):
    def __init__(self, in_size, out_size, activation):
        super().__init__()
        self.in_size, self.out_size, self.activation = in_size, out_size, activation
        self.block= nn.Identity()
        self.short= self.shortcut
    def forward(self, val):
        val= self.block(val)+self.short(val)
        return self.activation(val)
    
    @property
    def shortcut(self):
        return nn.Sequential(nn.Linear(self.in_size, self.out_size),
                             nn.BatchNorm1d(self.out_size))
```

```python
class Residual_Block(Residual_Base):
    def __init__(self, block_sizes, activation, p):
        super().__init__(block_sizes[0], block_sizes[-1], activation[-1])
        self.activ= activation
        self.block_sizes, self.p =block_sizes, p
        self.layer_sizes= list(zip(block_sizes, block_sizes[1:]))
        self.L= len(self.layer_sizes)
        self.hidden=[]
        self.hidden= self.get_block_layers()
        self.block= nn.Sequential(*self.hidden)

        
    def get_block_layers(self):

        for i,(input_size,output_size) in enumerate(self.layer_sizes):
            if i<self.L-1:
                self.hidden.append(nn.Linear(input_size, output_size, bias=False))
                self.hidden.append(nn.BatchNorm1d(output_size))
                self.hidden.append(self.activ[i])
                self.hidden.append(nn.Dropout(p=self.p))
            else:
                self.hidden.append(nn.Linear(input_size, output_size, bias=False))
                self.hidden.append(nn.BatchNorm1d(output_size))
        return self.hidden     
```

```python
class Gen(nn.Module):
    def __init__(self, in_size=3, n_layers=5, out_size=14943, p=0.0):
        
        super().__init__()
#         self.layer_sizes= [in_size, *[2**(2*i) for i in range(3, n_layers+2)],out_size]
        self.layer_sizes= [in_size, *[912 for _ in range(n_layers)], out_size]
        self.n_blocks= (len(self.layer_sizes)-2)//2
        self.block_sizes= [[self.layer_sizes[i+j*2] for i in range(3)] for j in range(self.n_blocks)]
        
        print(self.block_sizes)
        self.blocks= nn.Sequential(*[Residual_Block(self.block_sizes[i], [nn.ReLU(), nn.ReLU(inplace=True)], p) for i in range(self.n_blocks)])
    
        self.fc_layer= nn.Sequential(nn.Linear(self.layer_sizes[-2], self.layer_sizes[-1]),
                                     nn.Tanh())
       
    def forward(self, val):
        val= self.blocks(val)
        val= self.fc_layer(val)
        return val
```

```python
class Dis(nn.Module):
    def __init__(self, in_size=14943, n_layers=5, out_size=1, p=0.33, Wgan= False):
        
        super().__init__()
#         self.layer_sizes= [in_size, *[2**(2*i) for i in range(n_layers+1,2,-1)],out_size]
        self.layer_sizes= [in_size, *[912 for _ in range(n_layers)], out_size]
        
        self.n_blocks= (len(self.layer_sizes)-2)//2
        self.block_sizes= [[self.layer_sizes[i+j*2] for i in range(3)] for j in range(self.n_blocks)]
        print(self.block_sizes)
        self.blocks= nn.Sequential(*[Residual_Block(self.block_sizes[i], [nn.LeakyReLU(0.2), nn.LeakyReLU(0.2, inplace=True)], p) for i in range(self.n_blocks)])
    
        if not Wgan:
            self.fc_layer= nn.Sequential(nn.Linear(self.layer_sizes[-2], self.layer_sizes[-1]),
                                         nn.Sigmoid())
        else:
            self.fc_layer= nn.Linear(self.layer_sizes[-2], self.layer_sizes[-1])
                                         
            
    def forward(self, val):
        val= self.blocks(val)
        val= self.fc_layer(val)
        return val
```

```python
#Training Wasserstein Gan formulation
def WStein_Gan(train_loader, epochs=30):

    G= Gen_Sneha()
    D= Dis_Sneha(Wgan=True)
    
    D_iterations= 1
    Weight_clip= 0.01
    Gan_type= 'WStein_Gan'

    G_opt= torch.optim.RMSprop(G.parameters(), lr= 2e-4)
    D_opt= torch.optim.RMSprop(D.parameters(), lr= 2e-4)
    
    lossDict= {'Epoch':[],'D_loss':[], 'G_loss':[]}
    
    for epoch in range(epochs):
        D_lossCum=0
        G_lossCum=0
        for params, coords in train_loader:
            ## Discriminator Training
            for i in range(D_iterations):
                fake_coords= G(params+getRandomVector()).detach()
                D_loss= -(D(coords).mean() - D(fake_coords).mean())
                D_lossCum+=D_loss.item()
                D_opt.zero_grad()
                D_loss.backward(retain_graph= True)
                grad_vec= torch.cat([g.grad.contiguous().view(-1) for g in D.parameters()])
                print('Dis',grad_vec.min(), grad_vec.max(), grad_vec.mean(), grad_vec.std())
                D_opt.step()
                ## Clamping weights [-0.01, 0.01]
                for weight in D.parameters():
                    weight.detach().clamp_(-Weight_clip, Weight_clip)
            #D_lossCum=D_lossCum/D_iterations
            
            ##Generator training
            fake_coords= G(params) 
            G_loss= -D(fake_coords).mean()
            G_lossCum+=G_loss.item()
            G_opt.zero_grad()
            G_loss.backward()
            grad_vec= torch.cat([g.grad.contiguous().view(-1) for g in G.parameters()])
            print('Gen',grad_vec.min(), grad_vec.max(), grad_vec.mean(), grad_vec.std())
            G_opt.step()
            
        lossDict['Epoch'].append(epoch+1)
        lossDict['D_loss'].append(D_lossCum/batch_size)
        lossDict['G_loss'].append(G_lossCum/batch_size)
        #save_fake_imgs(G, Gan_type, epoch)  
        print('Epoch:{}, D_loss:{:.4f}, G_loss:{:.4f}'.format(epoch, D_loss, G_loss))
#     torch.save(G.state_dict(), './models/G_WGan.ckpt')
#     torch.save(D.state_dict(), './models/D_WGan.ckpt')
    plot(lossDict, Gan_type)
    
    return lossDict
   
        
```

```python
class SubHVPOperator(HVPOperator):
    def __init__(
        self,
        model,
        model2,
        dataloader,
        criterion,
        use_gpu=False,
        fp16=False,
        full_dataset=False,
        max_samples=256,):
        
        self.model2= model2
        super(SubHVPOperator, self).__init__(model,dataloader,criterion,use_gpu=False,fp16=False,full_dataset=False,max_samples=256)
        
    def _prepare_grad(self):
        """
        Compute gradient w.r.t loss over all parameters and vectorize
        """
        try:
            params, coords= next(self.dataloader_iter)
        except StopIteration:
            self.dataloader_iter= iter(self.dataloader)
            params, coords= next(self.dataloader_iter)
            
        if self.use_gpu:
            input = input.cuda()
            target = target.cuda()
        
        fake_coords= self.model2(params)
        loss= criterion(self.model(coords), 0.9*torch.ones_like(self.model(coords))) + criterion(self.model(fake_coords), torch.zeros_like(self.model(fake_coords)))

        grad_dict = torch.autograd.grad(
            loss, self.model.parameters(), create_graph=True
        )
        
        grad_vec = torch.cat([g.contiguous().view(-1) for g in grad_dict])
        
        grad_vec = utils.maybe_fp16(grad_vec, self.fp16)

        self.grad_vec = grad_vec
        return self.grad_vec
        
```
