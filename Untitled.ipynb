```python
class Geom_Dataset(Dataset):
    def __init__(self): #transform=None
        super().__init__()
        
        ## here you can add the file path
        self.data_name=pd.read_csv('./03_Kreuznapf/Experiments_1_without_Error_Term.csv', sep=';', decimal=',')
        self.x = torch.tensor(self.data_name[['F_press','fs','h_BT']].values, dtype = torch.float32)
        print("size of input is ",self.x.size())

        #Output
        disp_all = np.load("./03_Kreuznapf/disp_all_81000001.npy")
        disp_all = torch.tensor(disp_all, dtype=torch.float32)
        
        self.y = torch.hstack((disp_all[:,:,0],disp_all[:,:,1],disp_all[:,:,2]))
        self.y= self.transform()
        
        print("size of output is \n",self.y.size())
        print(self.y)
        self.len = self.x.shape[0]
        
    def transform(self):
        self.mean= self.y.mean()
        self.std= 3*self.y.std()
#         print(self.mean, self.std)
        return (self.y- self.mean)/self.std
    
    def __len__(self):
        return self.len
     
    def __getitem__(self, index):

        return self.x[index],self.y[index]
    
data = Geom_Dataset()
```

```python
val_size = 50
train_size = len(data) - val_size
batch_size = math.ceil(train_size/3)

train_ds, val_ds = random_split(data, [train_size, val_size])
print("the train size is {}, validation size is {} and batch size is {} ".format(len(train_ds), len(val_ds), batch_size))

train_loader = DataLoader(train_ds, batch_size, shuffle=True)
val_loader = DataLoader(val_ds, batch_size) 
```

```python
class Gen_Sneha(nn.Module):
    def __init__(self, in_size=3, h=5,s=150,outsize=14943, p=0):
        
        super().__init__()
        self.in_size = in_size
        self.outsize = outsize
        self.h = h
        self.L= h+1
        self.s = s
        self.hidden=nn.ModuleList()
        self.drop=nn.Dropout(p=p)
        self.Layers = [in_size, *[self.s*(i+1) for i in range(h)], outsize]
        print(self.Layers)
        
        for i,(input_size,output_size) in enumerate(zip(self.Layers, self.Layers[1:])):
            if i<self.L-1:
                self.hidden.append(nn.Linear(input_size, output_size, bias=False))
                self.hidden.append(nn.BatchNorm1d(output_size))
            else:
                self.hidden.append(nn.Linear(input_size, output_size, bias=False))
                
    def forward(self, activation):
        for(l,linear_transform) in enumerate(self.hidden):
            if l<self.L-1:
            
                activation = linear_transform(activation)
                activation=torch.relu(activation)
                activation=self.drop(activation)
            else:
                activation= linear_transform(activation)
                activation= torch.tanh(activation)
        return activation
```

```python
class Dis_Sneha(nn.Module):
    def __init__(self, in_size=14943, h=5,s=150,outsize=1, p=0.5, Wgan=False):
        
        super().__init__()
        self.in_size = in_size
        self.outsize = outsize
        self.h = h
        self.L= h+1
        self.s = s
        self.Wgan=Wgan
        self.hidden=nn.ModuleList()
        self.drop=nn.Dropout(p=p)
        self.Layers = [in_size, *[self.s*(i) for i in range(h,0,-1)], outsize]
        print(self.Layers)
    
        for i,(input_size,output_size) in enumerate(zip(self.Layers, self.Layers[1:])):
            if i<self.L-1:
                self.hidden.append(nn.Linear(input_size, output_size, bias=False))
                self.hidden.append(nn.BatchNorm1d(output_size))
            else:
                self.hidden.append(nn.Linear(input_size, output_size, bias=False))
                
    def forward(self, activation):
        for(l,linear_transform) in enumerate(self.hidden):
            if l<self.L-1:
            
                activation = linear_transform(activation)
                activation=nn.LeakyReLU(0.2)(activation)
                activation=self.drop(activation)
            else:
                activation= linear_transform(activation)
                if not self.Wgan:
                    activation= torch.sigmoid(activation)
        return activation
```

```python
def getRandomVector():
    rand_vect= torch.randn(batch_size, r_size)
    return rand_vect
```

```python
def plot(lossDict, ganType):
    fig = plt.figure(figsize=(10,8))


    plt.plot(lossDict['Epoch'],  lossDict['D_loss'], 'o-',label='D')
    plt.plot(lossDict['Epoch'],  lossDict['G_loss'], 'o-',label='G')

    ax = plt.gca()

    plt.ylabel('Cross-Entropy Loss')
    plt.xlabel('Epochs')
    plt.legend(fontsize=20, ncol=3) 

    plt.title('GANs Loss', pad=12)
#     plt.savefig("{}.png".format(ganType))
    plt.show()
```

```python
#Training Wasserstein Gan formulation
def WStein_Gan(train_loader, epochs=30):

    G= Gen_Sneha()
    D= Dis_Sneha(Wgan=True)
    
    D_iterations= 1
    Weight_clip= 0.01
    Gan_type= 'WStein_Gan'

    G_opt= torch.optim.RMSprop(G.parameters(), lr= 2e-4)
    D_opt= torch.optim.RMSprop(D.parameters(), lr= 2e-4)
    
    lossDict= {'Epoch':[],'D_loss':[], 'G_loss':[]}
    
    for epoch in range(epochs):
        D_lossCum=0
        G_lossCum=0
        for params, coords in train_loader:
            ## Discriminator Training
            for i in range(D_iterations):
                fake_coords= G(params+getRandomVector())
                D_loss= -(D(coords).mean() - D(fake_coords).mean())
                D_lossCum+=D_loss.item()
                D_opt.zero_grad()
                D_loss.backward(retain_graph= True)
                D_opt.step()
                ## Clamping weights [-0.01, 0.01]
                for weight in D.parameters():
                    weight.detach().clamp_(-Weight_clip, Weight_clip)
            #D_lossCum=D_lossCum/D_iterations
            
            ##Generator training
            fake_coords= G(params) 
            G_loss= -D(fake_coords).mean()
            G_lossCum+=G_loss.item()
            G_opt.zero_grad()
            G_loss.backward()
            G_opt.step()
            
        lossDict['Epoch'].append(epoch+1)
        lossDict['D_loss'].append(D_lossCum/batch_size)
        lossDict['G_loss'].append(G_lossCum/batch_size)
        #save_fake_imgs(G, Gan_type, epoch)  
        print('Epoch:{}, D_loss:{:.4f}, G_loss:{:.4f}'.format(epoch, D_loss, G_loss))
    torch.save(G.state_dict(), './models/G_WGan.ckpt')
    torch.save(D.state_dict(), './models/D_WGan.ckpt')
    plot(lossDict, Gan_type)
    
    return lossDict
   
        
```

```python
#Training non-saturating formulation
def NonSaturating_Gan(train_loader, epochs=200):

    G= Gen_Sneha()
    
    D= Dis_Sneha()
    D_iterations= 1

    Gan_type= 'NonSaturating_Gan'
    
 
    G_opt= torch.optim.Adam(G.parameters(), lr= 0.0271)
    D_opt= torch.optim.Adam(D.parameters(), lr= 0.0136)
    
    criterion= nn.BCELoss()
    lossDict= {'Epoch':[],'D_loss':[], 'G_loss':[]}
    
    for epoch in range(epochs):
        D_lossCum=0
        G_lossCum=0
        for params,coords in train_loader:
            fake_coords= G(params)
            for _ in range(D_iterations):
                ## Discriminator Training
                #D_loss= -(torch.log(D(real_imgs)).mean() + torch.log(1-D(fake_imgs)).mean())
                D_loss= criterion(D(coords), 0.9*torch.ones_like(D(coords))) + criterion(D(fake_coords), torch.zeros_like(D(fake_coords)))
                D_lossCum+=D_loss.item()
                D_opt.zero_grad()
                D_loss.backward(retain_graph= True)
                D_opt.step()
            
            ##Generator training
            #G_loss= -torch.log(D(fake_imgs)).mean()
            G_loss= criterion(D(fake_coords), torch.ones_like(D(fake_coords)))
            G_lossCum+=G_loss.item()
            G_opt.zero_grad()
            G_loss.backward()
            G_opt.step()
        
        lossDict['Epoch'].append(epoch)
        lossDict['D_loss'].append(D_lossCum/batch_size)
        lossDict['G_loss'].append(G_lossCum/batch_size)
        #save_fake_imgs(G, Gan_type, epoch)  
        print('Epoch:{}, D_loss:{:.4f}, G_loss:{:.4f}'.format(epoch, D_loss, G_loss))
    
    torch.save(G.state_dict(), './models/G_NonSaturating.ckpt')
    torch.save(D.state_dict(), './models/D_NonSaturating.ckpt')
    plot(lossDict, Gan_type)
    return lossDict
        
```
